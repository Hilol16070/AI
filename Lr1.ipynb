Решение:

```python
#Часть 1: Персональные данные
full_name = "Давлатов Хилол Илхомджонович"
group = "Группа: ПИН-бо-22-1"
specialty_code = "Шифр специальности: 09.03.03 Прикладная информатика"

# Вывод персональных данных
print(full_name)
print(group)
print(specialty_code)

# Часть 2: Перспективные языки программирования и ИТ-технологий
programming_languages = ["Python, C++"]
it_technologies = ["Машинное обучение, Наука о данных, Облачные вычисления"]

# Вывод перспективных языков и технологий
print("Перспективные языки программирования:")
for lang in programming_languages:
    print(f"- {lang}")

print("\nПерспективные ИТ-технологии:")
for tech in it_technologies:
    print(f"- {tech}")

# Часть 3: Увлечения и интересы
hobbies = "Чтение книг, путешествия, спортивные игры, программирование."
interests = "Интересуюсь новыми технологиями и саморазвитием."

# Вывод увлечений и интересов
print("Увлечения:")
print(hobbies)
print("\nИнтересы:")
print(interests)

```

2. Объявите требуемые переменные, присвойте им начальные значения (определите самостоятельно, значения какого типа могут принимать переменные), выведите на экран с использованием форматной строки значения переменных и результат вычисления выражения в соответствие с вариантом:

Решение:

```python
# Объявление переменных и присвоение начальных значений
w111 = 12      # целое число
bt = 8         # целое число
x = 5          # целое число
y = 3          # целое число
w = 4          # целое число

# Вычисление выражения Se = w111 + bt - x + y × w
Se = w111 + bt - x + y * w

# Вывод значений переменных и результата вычисления
print(f"Значение w111: {w111}")
print(f"Значение bt: {bt}")
print(f"Значение x: {x}")
print(f"Значение y: {y}")
print(f"Значение w: {w}")
print(f"Результат вычисления Se = w111 + bt - x + y × w: {Se}")
```

Объяснение:
1. Объявление переменных: В этом примере все переменные (w111, bt, x, y, w) имеют тип `int` и присваиваются целые значения.
2. Вычисление выражения: Выражение `Se` вычисляется с использованием заданной формулы.
3. Вывод: Все значения переменных и результат вычисления выводятся на экран с помощью форматных строк.
